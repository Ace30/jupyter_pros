```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
t = np.linspace(0, 40, 1000)

```

```python
d1 = 2.5*t
```

```python
d2 = 3*(t-5)
```

```python
fig, ax = plt.subplots()
plt.title('A Bank robber caught')
plt.xlabel('time (in mins)')
plt.ylabel('distance (in kms)')
ax.set_xlim([0, 40])
ax.set_ylim([0, 100])
ax.plot(t, d1, c = 'g')
ax.plot(t, d2, c = 'brown')
plt.axvline(x=30, c='b', linestyle = '--')
plt.axhline(y=75, c='b', linestyle = '--')






```
